```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
#import data in CSV file as DataFrame
df=pd.read_csv('organizations-100.csv', index_col=0)
df
```

```python
#getting info about our table
df.info()
```

```python
#retrieve first 15 elements of table 
df.head(15)
```

```python
#retrieve last 15 elements of table 
df.tail(20)
```

```python
#sort number of employees by ascending value 
df.sort_values(by='Number of employees', ascending=True, ignore_index=True)
```

```python
ad=df.groupby(['Country', 'Number of employees']).size().reset_index(name='Count')
ad
```

```python
print(df.groupby(['Country', 'Number of employees']).size().reset_index(name='Count').head(20))
```

```python
#visualize top 20 rows from grouped data using matplotlib 
plt.figure(figsize=(10, 6))
plt.scatter(df.groupby(['Country', 'Number of employees']).size().reset_index(name='Count').head(20)['Number of employees'], df.groupby(['Country', 'Number of employees']).size().reset_index(name='Count').head(20)['Country'], alpha=0.7)
plt.xlabel('Number of employees')
plt.ylabel('Country')
plt.grid(True)
plt.title('Number of Employees vs. Count by Country')
plt.show()
```
